```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
 
img1 = cv2.imread('images/kson.jpg')
img2 = np.zeros(img1.shape,np.uint8)
img2[:,:,0]=img1[:,:,2]
img2[:,:,1]=img1[:,:,1]
img2[:,:,2]=img1[:,:,0]
img3 = cv2.cvtColor(img1, cv2.COLOR_BGR2YUV)
 
plt.figure()
plt.subplot(2,1,1)
plt.imshow(img2)
plt.title('Original')
plt.subplot(2,1,2)
plt.imshow(img3)
plt.title('YUV')
plt.show()
```
